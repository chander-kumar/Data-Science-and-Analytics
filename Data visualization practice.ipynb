```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
plt.show()
```

```python
g=sns.FacetGrid(titanic, row='sex',hue='alone')
g=(g.map(plt.scatter,"age","fare").add_legend())
plt.show()
```

 ***Box Plot***

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
sns.set(style="darkgrid")
sns.boxplot(x='sex',y='tip',data=tips)
```

```python
sns.boxplot(x='tip',y='day',data=tips)
```

```python
sns.boxplot(x='tip',y='day',hue="smoker",data=tips)
```

```python
sns.boxplot(x='tip',y='day',hue="smoker",palette="Set1",dodge=False,data=tips)
```

```python
sns.boxplot(x='tip',y='day',hue="smoker",dodge=True,palette=['tab:blue', 'tab:red', 'tab:orange', 'tab:red'],data=tips)
```

```python
sns.boxplot(x='tip',y='day',showmeans=True, meanprops={"marker":"*","markersize":"12","markeredgecolor":"black"},data=tips)
plt.title("Tips ka chart",size=12,weight="bold")
```

```python
graph = sns.FacetGrid(tips, col ="sex",  hue ="day")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
# show the object
plt.show()
```

```python
graph = sns.FacetGrid(tips, row ='time', col ='smoker')
# map the above form facetgrid with some attributes
graph.map(plt.hist,'tip', bins = 15, color ='orange')
# show the object
plt.show()
```

```python
graph = sns.FacetGrid(tips, col ="sex",  row="smoker",hue="day")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
# show the object
plt.show()
  
```

```python
graph=sns.FacetGrid(tips,col="day")
graph.map(sns.boxplot,"tip")
plt.show()
```
